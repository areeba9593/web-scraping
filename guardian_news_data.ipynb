Cycling

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import time

# Define the base URL and number of pages to scrape
base_url = "https://www.theguardian.com/sport/cycling?page=24"
num_pages = 24

# Initialize lists to store data
list_titles_cyc = []
list_links_cyc = []
list_images_cyc = []
list_descriptions_cyc = []

for page_number in range(1, num_pages + 1):
    # Construct the URL for the current page
    url = base_url + str(page_number)
    
    # Request the page content
    response = requests.get(url)
    
    if response.status_code == 200:
        # Parse the content with BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all the news headlines
        coverpage_news = soup.find_all('h3')

        # Extract article titles, links, images, and descriptions
        for article in coverpage_news:
            title = article.get_text()
            list_titles_cyc.append(title)

            link = article.find('a')
            if link:
                list_links_cyc.append(link.get('href'))
            else:
                list_links_cyc.append("No link found")

            # Extract image if available
            image = article.find_previous('img')
            if image:
                list_images_cyc.append(image.get('src'))
            else:
                list_images_cyc.append("No image found")

            # Extract description if available
            description = article.find_next('p')
            if description:
                list_descriptions_cyc.append(description.get_text())
            else:
                list_descriptions_cyc.append("No description found")

        # Wait for a while before making the next request (to avoid hitting the server too hard)
        time.sleep(1)
    else:
        print(f"Failed to retrieve page {page_number}")

# Print the titles, links, images, and descriptions
print("Titles:")
print(list_titles_cyc)
print("\nLinks:")
print(list_links_cyc)
print("\nImages:")
print(list_images_cyc)
print("\nDescriptions:")
print(list_descriptions_cyc)
```

```python
data_cycling = {
    'Title': list_titles_cyc,
    'Link': list_links_cyc,
    'Image': list_images_cyc,
    'Description': list_descriptions_cyc
}


df_cycling = pd.DataFrame(data_cycling)


df_cycling['Sport'] = 'Cycling'

# Print the DataFrame
df_cycling
```

```python
import re

def extract_date(link):
    match = re.search(r'\d{4}/\w{3}/\d{2}', link)
    if match:
        return match.group()


complete_df['publication_date'] = complete_df['Link'].apply(extract_date)

complete_df
```
